```python
import pandas as pd
```

```python
#Reading the data set from the wikipedia website
url='https://en.wikipedia.org/wiki/List_of_countries_by_meat_consumption'
data=pd.read_html(url)
```

```python
pd.set_option('display.max_columns',10)
pd.set_option('display.width',1000)
df=data[1]
df
```

```python
import geoplot as gpt
import geopandas as gpd
```

```python
# the library geopandas consists of the built in data sets we will call them to get the shapes of each country
world_path=gpd.datasets.get_path('naturalearth_lowres')
world=gpd.read_file(world_path)
world.head()
```

```python
#we have to merge the geodata and table which consists the data of meat consumption
table=world.merge(df , how = 'left' , left_on= ['name'], right_on=['Country'])
print(table)
```

```python
#remove the null values
table=table.dropna(subset=['kg/person (2019)[13]'])
print(table)
```

```python
import folium
```

```python
my_map=folium.Map()
folium.Choropleth(
    geo_data= table,
    name='choropleth',
    data=table,
    columns=['Country','kg/person (2019)[13]'],
    key_on = 'feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Meat consumption in kg/person'
).add_to(my_map)
my_map.save('meat.html')
```
