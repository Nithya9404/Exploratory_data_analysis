```python
import matplotlib.pyplot as plt
```

```python
plt.barh(df['pH'],df['quality'])
plt.xlabel('quality')
plt.ylabel('pH')

#it is observed that the wine which has pH of 3.0 to 3.5 are of very good quality wine.The highest quality is 7.5 to 8.0
```

```python
import seaborn as sns
```

```python
cols=3
rows=3
num_cols=df.select_dtypes(exclude='object').columns
fig=plt.figure(figsize=(cols*5,rows*5))
for i,col in enumerate(num_cols):
  ax=fig.add_subplot(rows,cols,i+1)
  sns.histplot(x=df[col],ax=ax)

fig.tight_layout()
plt.show()
```

```python
print('Fixed_acidity: ',df['fixed acidity'].corr(df['pH']))
print('Volatile_acidity: ',df['volatile acidity'].corr(df['pH']))
print('Citric_acid: ',df['citric acid'].corr(df['pH']))
print('Residual sugar: ',df['residual sugar'].corr(df['pH']))
print('Chlorides: ',df['chlorides'].corr(df['pH']))
print('Free sulphur dioxide: ',df['free sulfur dioxide'].corr(df['pH']))
print('Total sulphur dioxide: ',df['total sulfur dioxide'].corr(df['pH']))
print('Density: ',df['density'].corr(df['pH']))
print('pH: ',df['pH'].corr(df['pH']))
```

```python
dataplot = sns.heatmap(df.corr(), cmap="YlGnBu", annot=True)
plt.show()

#ph has strong correlation with Volatile acidity and Free sulphur dioxide
# let us now visualize how Volatile acidity and Free sulphur dioxide contributes to the quality of wine
```

```python
plt.barh(df['volatile acidity'],df['quality'])
plt.xlabel('quality')
plt.ylabel('volatile acidity')
plt.show()

#it is observed that the volatile acidity of 0.0 to 1.3 can contribute to good quality of wine
```

```python
plt.bar(df['free sulfur dioxide'],df['quality'])
plt.xlabel('free sulphur dioxide')
plt.ylabel('quality')
plt.show()


#the quality of wine will be good if the free sulphur dioxide is less than 40
```
