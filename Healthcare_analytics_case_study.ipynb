```python
import seaborn as sns

sns.heatmap(data.corr(),cmap='YlGnBu')
```

```python
data_2020 = data[data['Year'] == 2020]
data_2020.head(4)
```

```python
gender_rate_mean = data_2020.groupby('Sex')['Malignant neoplasms (C00-C97)'].mean().reset_index(name='AverageRate')
gender_rate_mean
```

```python
age_wise_count = data_2020.groupby('Age Group')['Malignant neoplasms (C00-C97)'].mean().reset_index(name='Average_age_wise')
age_wise_count
```

```python
data_2021 = data[data['Year'] == 2021]
data_2021.head(4)
```

```python
gender_rate_mean_2021 = data_2021.groupby('Sex')['Malignant neoplasms (C00-C97)'].mean().reset_index(name='AverageRate')
gender_rate_mean_2021
```

```python
age_wise_count_2021 = data_2021.groupby('Age Group')['Malignant neoplasms (C00-C97)'].mean().reset_index(name='Average_age_wise')
age_wise_count_2021
```
