```python
data.groupby('Book_category')['Quantity'].count()
```

```python
# Matplotlib
table = pd.pivot_table(data=data,index='Book_category',values='Quantity')

# Barplot
plt.figure(figsize=(14,10))
plt.bar(table.index,table['Quantity'])
plt.xticks(rotation=70)
plt.xlabel('Category') 
plt.ylabel('Quantity') 
plt.show()
#maximum quantity of books in store is 18
#Contemporary
#Crime
#Cultural
#Poetry
```

```python
table = pd.pivot_table(data=data,index='Book_category',values='Price')

# Barplot
plt.figure(figsize=(25,10))
plt.bar(table.index,table['Price'])
plt.xticks(rotation=70)
plt.xlabel('Category') 
plt.ylabel('Price') 
plt.show()

#observation:
#The category Suspence has highest cost is  i.e almost about 58
#The category Crime has lowest cost i.e., almost 10
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
# A regular scatter plot
plt.scatter(x=data["Book_category"], y=data["Star_rating"])

plt.rcParams['figure.figsize'] = [25, 10]
plt.xticks(rotation=70)
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()

#observation:
#The categories like 'Cultural','Psychology','Humor','Sports and Games','Parenting','Paramonial' do not have good star rating
```

```python
def f(x, y):
 return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(1, 5, 100)
y = np.linspace(2, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, colors='black');
```

```python
#basic operations in numpy
x = np.array([[9,15,23], [2, 3, 4]])
y = np.array([[2,9,0], [1,9,10]])

add=np.add(x,y)
print("Addition",add)
print()
sub=np.subtract(x,y)
print("Subtraction",sub)
print()
mul=np.multiply(x,y)
print("Multiplication",mul)
print()
div=np.remainder(x,y)
print("Division",div)
```
